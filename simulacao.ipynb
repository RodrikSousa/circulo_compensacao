```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

eleicao = 'eleicoes\AR_2022_Globais_REPEU.xlsx'

sheet_nacional = 'AR_2022_Distrito_Reg.Autónoma' 
df_nac = pd.read_excel(eleicao, sheet_name=sheet_nacional, skiprows = 3, nrows = 21)
df_nac.head()
```

```python
sheet_int = 'AR_2022_Global' 
df_int = pd.read_excel(eleicao, sheet_name=sheet_int, skiprows = 3, nrows = 5)
df_int.head()
```

```python
df_total = pd.concat([df_nac.loc[df_nac["código"]!=500000], 
                      df_int.loc[df_int["código"]==800000], 
                      df_int.loc[df_int["código"]==900000]])   
df_total.head()
```

```python
df_base = pd.concat([df_total.iloc[:,0:2],  df_total.iloc[:,6:9], df_total.iloc[:,13]], axis = 1).reset_index(drop = True)
df_base
```

```python
df_partidos = pd.concat([df_total.iloc[:,0:2], df_total.iloc[:,14:], df_total.iloc[:,9:13]], axis = 1).fillna(0)
df_partidos.head()
```

```python
import janitor

df_partidos_treat = df_partidos.pivot_longer(
                        index = ['código','nome do território']
                        , names_to = ["partido", "drop1", "drop2"]
                        , values_to = ["votos", "% votos", "mandatos"]
                        , names_pattern = ["^brancos|^nulos|[A-Z]", "% brancos|% nulos|% votantes.*", "mandatos.*"]
                    ).drop(columns=['drop1', 'drop2'])
df_partidos_treat.head()
```

```python
df_base["eleitores por mandato"] = df_base["inscritos"]/df_base["mandatos"]
df_base.sort_values("eleitores por mandato")
```

```python
pd.options.mode.chained_assignment = None
def reduzir(df, tam_circ_comp, min_circ):
    i = 0
    df = df.copy(deep=True)
    df["mandatos"] = df["mandatos"].apply(lambda x: max(x, min_circ))
    while i < tam_circ_comp:
        df["eleitores por mandato"] = df["inscritos"]/df["mandatos"]
        df["atingiu minimo"] = df["mandatos"].apply(lambda x: x == min_circ)
        df.sort_values(["atingiu minimo","eleitores por mandato"], inplace = True, ignore_index=True)
        df["mandatos"][0] -= 1
        i += 1
    return df

tamanho_circulo_compensacao = 0
tamanho_circulo_minimo = 2

df_reduzido = reduzir(df_base, tamanho_circulo_compensacao, tamanho_circulo_minimo)
df_reduzido.sort_values(["eleitores por mandato"])
```

```python
df_votos_nacional = df_partidos_treat[~df_partidos_treat.partido.isin(["brancos", "nulos"])].groupby("partido", as_index =False)[['votos', 'mandatos']].sum()
df_votos_nacional["%votos"] = df_votos_nacional["votos"]/sum(df_votos_nacional["votos"])
df_votos_nacional["%mandatos"] = df_votos_nacional["mandatos"]/sum(df_votos_nacional["mandatos"])
df_votos_nacional["desvio"] = abs(df_votos_nacional["%votos"]-df_votos_nacional["%mandatos"])
df_votos_nacional
```

```python
soma_dos_desvios = sum(df_votos_nacional["desvio"])
soma_dos_desvios
```
